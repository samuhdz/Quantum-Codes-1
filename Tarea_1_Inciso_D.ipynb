```python
import numpy as np
import scipy as sci
from scipy.constants import h,c,k
import matplotlib.pyplot as plt

def Plancks_law(T, l):
    return ((2 * np.pi * h * c**2) / (l**5)) * (1 / (np.exp((h * c) / (k * T * l)) - 1))

lambda_1=np.linspace(0.000000000001,200*1e3,200000)
lambda_2=np.linspace(0.000000000001,2000,200000)
lambda_3=np.linspace(0.000000000001,200,200000)
Ts=np.array([100,5800,120000])

I1=Plancks_law(100,lambda_1*1e-9)
I2=Plancks_law(5800,lambda_2*1e-9)
I3=Plancks_law(120000,lambda_3*1e-9)

fig, axs =plt.subplots(3,1, figsize=(8,12))
axs[0].plot(lambda_1,I1, label='T=100K', color='r')
axs[0].legend()
axs[0].grid(True)
axs[1].plot(lambda_2,I2,label='T=5800K', color='g')
axs[1].set_ylabel('Intensidad (KJ/nm)')
axs[1].legend()
axs[1].grid(True)
axs[2].plot(lambda_3,I3, label='T=120000K', color='b')
axs[2].set_xlabel('Longitud de onda (nm)')
axs[2].legend()
axs[2].grid(True)


plt.tight_layout()
plt.show()
```

```python
#Para hallar los máximos lo hacemos como en IP
l1=(lambda_1*1e-9)[np.argmax(I1)]*1e9
l2=(lambda_2*1e-9)[np.argmax(I2)]*1e9
l3=(lambda_3*1e-9)[np.argmax(I3)]*1e9

print("El máximo para T=100K es " + str(l1) + 
      'nm, para T=5800K es ' + str(l2)+ 'nm  \n' + 'y para T=120000K es ' + str(l3) + 'nm')
```
